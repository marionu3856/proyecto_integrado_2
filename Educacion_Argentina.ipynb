```python
# prompt: Armar grafico de barras con Promedio general de cada materia, ordenados de mayor a menor, agregar las faltas como valor constrastable en dos ejes. Agregar las leyendas que ayuden a la interpretacin.

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data (replace with your actual data)
data = {'Materia': ['Matemticas', 'Fsica', 'Qumica', 'Historia', 'Literatura'],
        'Promedio': [8.5, 7.2, 9.1, 6.8, 8.0],
        'Faltas': [2, 5, 1, 7, 3]}
df = pd.DataFrame(data)

# Sort by promedio
df_sorted = df.sort_values('Promedio', ascending=False)


# Create the bar plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Promedio
ax1.bar(df_sorted['Materia'], df_sorted['Promedio'], color='skyblue', label='Promedio')
ax1.set_xlabel('Materia')
ax1.set_ylabel('Promedio', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_title('Promedio General de Materias (Ordenado) y Faltas')

# Create a second y-axis for Faltas
ax2 = ax1.twinx()
ax2.plot(df_sorted['Materia'], df_sorted['Faltas'], marker='o', color='red', label='Faltas')
ax2.set_ylabel('Nmero de Faltas', color='red')
ax2.tick_params(axis='y', labelcolor='red')


# Add labels and annotations for Faltas
for i, falta in enumerate(df_sorted['Faltas']):
    ax2.annotate(falta, (df_sorted['Materia'][i], falta), textcoords="offset points", xytext=(0,5), ha='center')

# Add legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')


plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
